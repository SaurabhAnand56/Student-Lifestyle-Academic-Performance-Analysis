# Feature Engineering

```python
df["total_screen_time"] = (
    df["social_media_hours"] +
    df["gaming_hours"] +
    df["streaming_hours"]
)
```

### Insight
A new feature "total_screen_time" was created to represent the
total amount of time students spend on digital entertainment.

## Univariate Analysis
### Age Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(df["age"], bins=10)
plt.title("Age Distribution")
plt.show()
```

### Insight
Most students fall within the typical academic age range.
The distribution appears fairly balanced with no extreme age outliers,
indicating the dataset represents a normal student population.

### Gender Distribution

```python
sns.countplot(x="gender", data=df)
plt.title("Gender Distribution")
```

### Mental Health Status

```python
sns.countplot(x="mental_health_status", data=df)
```

## Study Behavior Analysis

### Study Hours Distribution

```python
sns.histplot(df["study_hours_per_day"], bins=20)
```

Only a small number of students study more than 7 hours,

## Bivariate Analysis
### Study Hours vs Exam Score

```python
sns.scatterplot(
x="study_hours_per_day",
y="final_exam_score",
data=df
)
```

### Insight
There is a positive relationship between study hours and exam scores.
Students who dedicate more time to studying tend to achieve higher
academic performance.

### Smartphone Usage vs Exam Score

```python
sns.scatterplot(
x="smartphone_usage_hours",
y="final_exam_score",
data=df
)
```

### Attendance vs Exam Score

```python
sns.scatterplot(
x="class_attendance_percent",
y="final_exam_score",
data=df
)
```

### Insight
Class attendance shows a strong positive relationship with exam scores.
Students who attend more classes generally perform better academically.

### Sleep vs Exam Score

```python
sns.scatterplot(
x="sleep_hours",
y="final_exam_score",
data=df
)
```

### Insight
Students who maintain balanced sleep patterns appear to perform
better academically, suggesting that adequate rest contributes
to better cognitive performance.

## Categorical Analysis
### Gender vs Exam Score

```python
sns.boxplot(
x="gender",
y="final_exam_score",
data=df
)
```

### Insight
The distribution of exam scores appears relatively similar across genders,
indicating that gender may not be a major factor influencing academic
performance in this dataset.

### Mental Health vs Exam Score

```python
sns.boxplot(
x="mental_health_status",
y="final_exam_score",
data=df
)
```

### Insight
Students with better mental health status tend to achieve higher
exam scores, highlighting the importance of psychological well-being
for academic success.

```python
sns.boxplot(x=df["smartphone_usage_hours"])
```

### Insight
A few students show extremely high smartphone usage,
which may indicate potential digital overuse.

## Correlation Analysis

```python
plt.figure(figsize=(12,8))
sns.heatmap(
df.corr(numeric_only=True),
annot=True,
cmap="coolwarm"
)
```

### Insight
The correlation heatmap shows that study hours, class attendance,
and assignment completion have strong positive relationships with
final exam scores.

In contrast, excessive digital usage variables show weaker or
slightly negative relationships with academic performance.

### Target Variable Analysis

```python
df.corr(numeric_only=True)["final_exam_score"].sort_values(ascending=False)
```

### Insight
Study hours, attendance, and motivation level appear to
be the most influential factors affecting exam scores.

### Top vs Low Performers

```python
top_students = df[df["final_exam_score"] > 90]
top_students.describe()
```

### Insight
Top-performing students generally have higher study hours,
better attendance, and balanced digital usage.

### Low Students

```python
low_students = df[df["final_exam_score"] < 60]
low_students.describe()
```

### Insight
Low-performing students often show lower study hours
and potentially higher screen time.

## Key findings from the analysis:

1. Students studying more than 6 hours per day tend to achieve higher scores.
2. Class attendance has strong influence on academic performance.
3. High screen time may negatively impact study behavior.
4. Balanced sleep patterns appear beneficial for academic success.
5. Mental health status plays an important role in student performance.
